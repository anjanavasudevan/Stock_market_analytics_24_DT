## Question 1

```python
# dataset:
# Set the date
end = date.today()
print(f'Year = {end.year}; month= {end.month}; day={end.day}')

start = date(year=end.year-25, month=end.month, day=end.day) # Upto 10 yrs of data
print(f'Period for indexes: {start} to {end} ')

# Data
# source: https://fred.stlouisfed.org/series/GDPC1
gdpc_data = pdr.DataReader("GDPC1", "fred", start=start)

gdpc_data.tail(10)
```

```python
# calculate YoY growth
gdpc_data["YoY_growth"] = gdpc_data["GDPC1"]/gdpc_data["GDPC1"].shift(4)-1 # Org-diff/org formula rearranged
gdpc_data.tail()
```

```python
# Avg value:
new_df = gdpc_data.tail(4).copy()
display(new_df.describe())
display(new_df)
```

```python
## ANSWER - 2.5%
```

```python
# Index data:
start = date(2019,4,9)
end = date(2024, 4, 9)

sp500 = pdr.get_data_stooq('^SPX', start, end)
display(sp500.head(10))
```

```python
ipc = pdr.get_data_stooq('^IPC', start, end)
display(ipc.head(10))
```

```python
sp500_range = sp500[sp500.index.isin([start, end])].copy()
ipc_range = ipc[ipc.index.isin([start, end])].copy()

display(sp500_range)
display(ipc_range)

sp500_growth = sp500_range["Close"]/sp500_range["Close"].shift(-1) - 1
ipc_growth = ipc_range["Close"]/ipc_range["Close"].shift(-1) - 1
display(sp500_growth)
display(ipc_growth)
```

## Question 4

```python
# Get stock data for the companies
SR = yf.download("2222.SR", start="2023-01-01", end="2023-12-31", actions=True)
display(SR.head())

BRK = yf.download("BRK-B", start="2023-01-01", end="2023-12-31", actions=True)
AAPL = yf.download("AAPL", start="2023-01-01", end="2023-12-31", actions=True)
MSFT = yf.download("MSFT", start="2023-01-01", end="2023-12-31", actions=True)
GOOG = yf.download("GOOG", start="2023-01-01", end="2023-12-31", actions=True)
JPM = yf.download("JPM", start="2023-01-01", end="2023-12-31", actions=True)
```

```python
# Calculate max and min adj close
ticker_stats = {"Ticker": [], "Max": [], "Min": []}

ticker_stats["Ticker"].append("2222.SR")
ticker_stats["Max"].append(SR["Adj Close"].max())
ticker_stats["Min"].append(SR["Adj Close"].min())

ticker_stats["Ticker"].append("BRK-B")
ticker_stats["Max"].append(BRK["Adj Close"].max())
ticker_stats["Min"].append(BRK["Adj Close"].min())

ticker_stats["Ticker"].append("AAPL")
ticker_stats["Max"].append(AAPL["Adj Close"].max())
ticker_stats["Min"].append(AAPL["Adj Close"].min())

ticker_stats["Ticker"].append("MSFT")
ticker_stats["Max"].append(MSFT["Adj Close"].max())
ticker_stats["Min"].append(MSFT["Adj Close"].min())

ticker_stats["Ticker"].append("GOOG")
ticker_stats["Max"].append(GOOG["Adj Close"].max())
ticker_stats["Min"].append(GOOG["Adj Close"].min())

ticker_stats["Ticker"].append("JPM")
ticker_stats["Max"].append(JPM["Adj Close"].max())
ticker_stats["Min"].append(JPM["Adj Close"].min())

# convert to df
ticker_stats_df = pd.DataFrame(ticker_stats)
display(ticker_stats_df)
```

```python
ticker_stats_df["ratio"] = (ticker_stats_df["Max"]-ticker_stats_df["Min"])/ticker_stats_df["Max"]

ticker_df = ticker_stats_df.copy()
for column in ["Max", "Min", "ratio"]:
  ticker_df[column] = ticker_df[column].round(2)

display(ticker_df)
```

## Question 5

```python
ticker_data = {"2222.SR": SR, "BRK-B": BRK, "AAPL": AAPL, "MSFT": MSFT, "GOOG": GOOG, "JPM": JPM}
yield_df = {"Ticker": [], "Yield": []}

for ticker in ticker_data.keys():
  td = ticker_data[ticker].copy()
  ticker_yield = td["Dividends"].sum()/td.tail(1)["Adj Close"].values[0] # Assumes that the data is sorted from less recent to more recent
  yield_df["Ticker"].append(ticker)
  yield_df["Yield"].append(ticker_yield)

ticker_yield_df = pd.DataFrame(yield_df)
display(ticker_yield_df)
```

```python
# Answer
max_yield = round(ticker_yield_df["Yield"].max()*100, 1)
print(max_yield)
```
